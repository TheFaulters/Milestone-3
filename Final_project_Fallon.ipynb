```python
# get dummies for the loan status column
status_dummies = pd.get_dummies(data['loan_status'])
```

```python
data = pd.concat((data, status_dummies), axis = 1)
```

```python
col_names = status_dummies.columns.values

count_subframe = data[['year', 'int_rate', 'annual_inc', col_names[0], col_names[1], col_names[2], col_names[3], col_names[4],
                      col_names[5], col_names[6], col_names[7], col_names[8]]]

count_subframe.head(n=2)
count_sample = count_subframe.sample(frac = .10)
```

```python
#break into 3 groups > PAID, DELINQUENT, DEFAULT

count_sample.loc[:, 'P'] = count_sample[['Fully Paid', 
                                         'Does not meet the credit policy. Status:Fully Paid']].apply(lambda x : sum(x), axis = 1)
count_sample.loc[:, 'Del'] = count_sample[['Late (31-120 days)', 
                                            'Current',
                                            'In Grace Period', 
                                            'Late (16-30 days)']].apply(lambda x : sum(x), axis = 1)
count_sample.loc[:, 'Def'] = count_sample[['Charged Off', 
                                            'Default', 
                                            'Does not meet the credit policy. Status:Charged Off']].apply(lambda x : sum(x), axis = 1)

```

```python
count_sample.groupby('year')

count_sample.head(n=2)
#keeping other variables in for graphing later (did this late at night, will refine later)
count_graph = count_sample[['year', 'int_rate', 'annual_inc', 'P', 'Del', 'Def']]
count_graph.head(n=2)
```

```python
#group by years and sum each category for each year, replace abbreviations with more clear column names
dic = {'P': 'Paid', 'Del' : 'Delinquent', 'Def' : 'Default'}
count_graph.rename(columns = dic, inplace = True)
count_graph_group = count_graph.groupby('year').sum()
```

```python
fig, ax = plt.subplots(figsize = (15,8))
labels = list(count_graph_group.index.values)
labels.insert(0,' ')
x1 = [-0.3, 0.7, 1.7, 2.7, 3.7]
x2 = [0, 1, 2, 3, 4]
x3 = [.3, 1.3, 2.3, 3.3, 4.3]
x = count_graph_group.index
y1 = count_graph_group['Paid']
y2 = count_graph_group['Delinquent']
y3 = count_graph_group['Default']
w = 0.2
ax.bar(x1, y1, width = .3, color = 'r', label = 'Paid', align = 'center')
ax.bar(x2, y2, width = .3, color = 'b', label = 'Delinquent', align = 'center')
ax.bar(x3, y3, width = .3, color = 'g', label = 'Default', align = 'center')
ax.set_xticklabels(labels, fontsize = 20)
ax.set_xlim(-1,5)
ax.legend()
ax.set_title('Loan status over time', fontsize = 22)
plt.savefig('Loan status over time.png');
```
